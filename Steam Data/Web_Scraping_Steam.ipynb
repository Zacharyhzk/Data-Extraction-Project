```python
import urllib
import urllib.request
import time
from datetime import datetime
from bs4 import BeautifulSoup
import pandas as pd
import re
import requests
```

```python
# Try to collect game details and url links of topseller games in Steam
gamesNameList = []
releaseTimeList = []
gamePriceList = []
titleURLList = []
applyIdList = []
 
url = 'https://store.steampowered.com/search/?filter=topsellers'
headers = {"Accept-Language": "en-US, en;q=0.5"}
results = requests.get(url, headers=headers)
#Beautiful Soup is a Python library for pulling data out of HTML and XML files. 
soup = BeautifulSoup(results.text, "html.parser")

gamesName = soup.find_all("span", {"class": "title"})
releaseTime = soup.find_all("div", {"class": "search_released"})
gamesPrice = soup.find_all("div", {"class": "search_price"})
titleURL = soup.find("div", {"id": "search_resultsRows"}).find_all("a")
 
for item in gamesName:
    gamesNameList.append(item.get_text().strip())                 
for item in releaseTime:
    releaseTimeList.append(item.get_text().strip())
for item in gamesPrice:
    gamePriceList.append(item.get_text().strip())
for item in titleURL:
    titleURLList.append(item['href'])
    # add apply id 
    match = re.search(r'\b([0-9]+)\b', item['href'])
    applyIdList.append(match.group(0))

# all_games_list.extend(games)

# print(gamesNameList)
# print(searchTimeList)
# print(gamePriceList)
# print(titleURLList)
gamesDetail = pd.DataFrame({
    'gameName': gamesNameList,
    'releaseTime': releaseTimeList,
    'price': gamePriceList,
    'titleURL': titleURLList,
    'applyId': applyIdList
  })
gamesDetail.head(10)               
# print(all_games_list)
```

```python
gamesDetail.to_csv('game_Details.csv', index=False, encoding='utf-8')
```

```python
gameNameList = []
developerList = []
gameTypesList = []
gameDescriptionList = []

for url in gamesDetail.titleURL:
    headers = {"Accept-Language": "en-US, en;q=0.5"}
    results = requests.get(url, headers=headers)
    #Beautiful Soup is a Python library for pulling data out of HTML and XML files. 
    soup = BeautifulSoup(results.text, "html.parser")
    # Try to collect game details and url links of topseller games in Steam
    game_title = []
    game_Types = []
    
    print(url)
    
    game_name = soup.find("div", {"id": "appHubAppName"})
    try:
        gameName = game_name.get_text().strip() 
#         if type(game_name) == str else soup.find("h2", {"class": "pageheader"}).get_text().strip()
        developer = soup.find("div", {"id": "developers_list"}).find("a").get_text().strip()
        description = soup.find("div", {"class": "game_description_snippet"}).get_text().strip()
    except AttributeError:
        gameName = " "
        developer = " "
        description = " "
    game_title.append(gameName)
    gameTypes = soup.find_all("a", {"class": "app_tag"})
    for item in gameTypes:
        if type(item) == str: break
        gameType = item.get_text().strip()
        game_Types.append(gameType)
    
    gameNameList.append(game_title)
    developerList.append(developer)
    gameTypesList.append(game_Types)
    gameDescriptionList.append(description)

#     print(gameNameList)
#     print(developerList)
#     print(gameTypesList)
#     print(gameDescriptionList)
#     print(len(gameNameList))
#     print(len(developerList))
#     print(len(gameTypesList))
#     print(len(gameDescriptionList))

gameDetailFromURL = pd.DataFrame({
    'gameName': gameNameList,
    'developer': developerList,
    'gameType': gameTypesList,
    'gameDescription': gameDescriptionList
  })
gameDetailFromURL.head(100)

```

```python
# get review from steam review api
#https://steamcommunity.com/app/applyId/reviews/?browsefilter=toprated&snr=1_5_100010_

reviewContentsList = []
ratesContentList = []
gameTitleList = []

for applyId in gamesDetail.applyId:
    # special case
    if (str(applyId) == "1708091"):
        applyId =  "1240440"
    if (str(applyId) == "1746860"):
        applyId = "489830"
    if (str(applyId) == "1445990"):
        applyId = "1172470"
    if (str(applyId) == "1766730"):
        applyId = "730"
    if (str(applyId) == "1348661"):
        applyId = "394360"
    if (str(applyId) == "1656360"):
        applyId = "1085660" 
    if (str(applyId) == "1314561"):
        applyId = "1085660" 
    if (str(applyId) == "1289670"):
        applyId = "1172470"
    if (str(applyId) == "1333220"):
        applyId = "1286830"
     
    reviewContents = []
    ratesContents = []
    gameTitles = []
    url = "https://steamcommunity.com/app/" + str(applyId) + "/reviews/?browsefilter=toprated&snr=1_5_100010_"
    print(url)
    headers = {"Accept-Language": "en-US, en;q=0.5"}
    results = requests.get(url, headers=headers)
    #Beautiful Soup is a Python library for pulling data out of HTML and XML files. 
    soup = BeautifulSoup(results.text, "html.parser")
    # Try to get the game title 
    title = soup.find("div", {"class": "apphub_AppName"})
    try:
        gameTitle = title.get_text()
    except AttributeError:
        gameTitle = " "
#     print(gameTitle)
    # Try to collect all the review box in steam
    review_box = soup.find_all("div", {"class": "apphub_CardContentMain"})
    for item in review_box:
        # Try to collect reviews of game in Steam
        review_Contents = item.find_all("div", {"class": "apphub_CardTextContent"})
        for review_Content in review_Contents:
            reviewContent = review_Content.get_text().strip()
            reviewContents.append(reviewContent)     
        # Try to collect rate from user in the reviews
        rates = item.find_all("div", {"class": "reviewInfo"})
        for rate in rates:
            ratesContents.append(rate.find("div", {"class": "title"}).get_text())
        gameTitles.append(gameTitle)
        
    reviewContentsList += reviewContents
    ratesContentList += ratesContents
    gameTitleList += gameTitles
#     print(len(reviewContentsList))
#     print(len(ratesContentList)) 
#     print(len(gameTitleList))
#     print(gameTitleList)
    
reviewsDetail = pd.DataFrame({
    'gameTitle': gameTitleList,
    'Reviews':reviewContentsList,
    'Rate':ratesContentList,
  })
reviewsDetail.head(100)
```

```python
reviewsDetail.to_csv('review_Details.csv', index=False, encoding='utf-8')
```
