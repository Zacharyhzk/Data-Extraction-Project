```python
# Get a list of all urls to the IMDb Most Popular Movies
#urls = https://www.imdb.com/chart/moviemeter/?ref_=nv_mv_mpm
# Import Libraries 
import requests
from requests import get
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np

import re
import requests
import time
import csv
from selenium import webdriver


url="https://www.imdb.com/chart/moviemeter/?ref_=nv_mv_mpm"
baseURL = "https://www.imdb.com"
# Initiate data storage
titleURLs = list()
titleURLsWithReviews = list()

headers = {"Accept-Language": "en-US, en;q=0.5"}
results = requests.get(url, headers=headers)
#Beautiful Soup is a Python library for pulling data out of HTML and XML files. 
soup = BeautifulSoup(results.text, "html.parser")
title_urlAll = soup.find_all("td",{"class": "titleColumn"})
for titleURL in title_urlAll:
    titleURLs.append(baseURL + titleURL.find("a").get("href"))

# title_urlAllWithReviews = soup.find_all("tbody",{"class": "lister-list"})
title_urlAllWithReviews = soup.find_all("tr")
# print(title_urlAllWithReviews)
for titleURLWithReviews in title_urlAllWithReviews:
#     print(titleURLWithReviews.find("strong"))
    if titleURLWithReviews.find("strong") is not None:
        titleURLsWithReviews.append(baseURL + titleURLWithReviews.find("a").get("href"))
    else:
        titleURLsWithReviews.append(" ")
titleURLsWithReviews.pop(0)  # to pop the first none value

movies = pd.DataFrame({
    'titleURLs': titleURLs,
    'titleURLsWithReviews': titleURLsWithReviews,
})


#add dataframe to csv name 'movies.csv'
# movies.to_csv('mp_movies.csv', index=False, encoding='utf-8')

#display ‘movies.xlsx’
movies.head(1000)
```

```python
movieTitles = []
directors =[]
actorNamesList = []
movieTypesList = []
movieRateList = []
movieYearList = []
for i in movies.titleURLsWithReviews:
    if i == " ":continue
#     print(i)
    actorNames = []
    movieTypes = []
    headers = {"Accept-Language": "en-US, en;q=0.5"}
    results = requests.get(i, headers=headers)
    #Beautiful Soup is a Python library for pulling data out of HTML and XML files. 
    soup = BeautifulSoup(results.text, "html.parser")
    movieTitle = soup.find("h1",{"class": re.compile("TitleHeader__TitleText*")}).get_text()
    director = soup.find("a",{"class": re.compile("ipc-metadata-list-item__list-content-item ipc-metadata-list-item__list-content-item--link")}).get_text()
    runningTime = soup.find("ul",{"class": re.compile("baseAlt")}).find("li")
    actor_Names = soup.find_all("a",{"class": re.compile("StyledComponents__ActorName*")})
    movie_types = soup.find_all("a",{"class": re.compile("ipc-chip--on-baseAlt*")})
    movieRate = soup.find("span",{"class": re.compile("AggregateRatingButton__Rating*")}).get_text()
    movieYear = soup.find("a",{"class": re.compile("TitleBlockMetaData__StyledTextLink*")}).get_text() 
    for item in movie_types:
        movieType = item.find("span",{"class": re.compile("ipc-chip__text")}).get_text()
        movieTypes.append(movieType)
    for item in actor_Names:
        if type(item) == str: break
        actorName = item.get_text() 
        actorNames.append(actorName)
    movieTitles.append(movieTitle)
    directors.append(director)
    actorNamesList.append(actorNames)
    movieTypesList.append(movieTypes)
    movieRateList.append(movieRate)
    movieYearList.append(movieYear)
    
moviesDetail = pd.DataFrame({
    'movieTitle': movieTitles,
    'director':directors,
    'actorNamesList':actorNamesList,
    'movieTypesList':movieTypesList,
    'movieRateList':movieRateList,
    'movieYearList':movieYearList,
  })
moviesDetail.head(100)
```

```python
moviesDetail.to_csv('movie_Details.csv', index=False, encoding='utf-8')
```

```python
def generate_movie_review_list_link(urls):
    movie_urls = []
    for item in urls:
        if item == " ":continue 
        for i in range(1,4):
            #       filter spoiler
            tmp = ''
            tmp = item+"reviews?spoiler=hide&sort=helpfulnessScore&dir=desc&ratingFilter="+str(i)
            movie_urls.append(tmp)
        for i in range(8,11):
            tmp = ''
            tmp = item+"reviews?spoiler=hide&sort=helpfulnessScore&dir=desc&ratingFilter="+str(i)
            movie_urls.append(tmp)
    return movie_urls

movie_reviews_urls = generate_movie_review_list_link(movies.titleURLsWithReviews)
```

```python
#get review information from selected movies
def get_imdb_movie_review(url,times):
    URL = url
    print(URL)
    response = requests.get(URL)
    if response.status_code == 200:
        original_html = response.text
        original_soup = BeautifulSoup(original_html,'lxml')
        path = os.getcwd() + "\\chromedriver2_win32" + "\\chromedriver.exe" 
        
        chrome_options = Options() 
        chrome_options.add_argument("--headless") 
        chrome_options.add_argument("--window-size=1920x1080") 
        
        chrome_driver = os.getcwd() + "\\chromedriver2_win32" + "\\chromedriver.exe" 
        service = Service(chrome_driver)
        browser = webdriver.Chrome(service=service, options=chrome_options) ### headless browser - faster
        
        browser.get(URL)
        time.sleep(3)

        # click load more button
        for i in range(times):
            temp_response = browser.page_source
            temp_html = temp_response
            temp_soup = BeautifulSoup(temp_html,'lxml')
            load_more = temp_soup.select('.ipl-load-more__button')
            #have load_more btn
            if(len(load_more) == 1):
                button_load_more = browser.find_element(By.CLASS_NAME,"ipl-load-more__button")
                button_is_or_not_visible = browser.find_element(By.CLASS_NAME,"ipl-load-more__button").is_displayed()
                if(button_is_or_not_visible is True):       #once load more button is invisible then stop clicking
                    button_load_more.click()
                    time.sleep(3)
                else:
                    break
            else:
                break

        final_response = browser.page_source
        html = final_response
        soup = BeautifulSoup(html,'lxml')

        reviews = soup.select('.review-container')
        

        for review in reviews:
            rate = review.select_one('.rating-other-user-rating')
            review_label.append('Negative' if rate.select_one('span').text in ['1','2','3'] else 'Positive')
            content = review.select_one('.content')
            user_review = content.select('.text.show-more__control')             
            review_content.append(user_review[0].text)
            time.sleep(0.01)

        browser.close()      # close

        print(f'{len(reviews)} pieces of reviews have been scraped.')
    else:
        print("Error when request URL")
        

review_content = []
review_label = []
for url in movie_reviews_urls:
#each page will display 25 pieces of review, then click 'load more' button 3 times will get 100 pcs of review.
    get_imdb_movie_review(url,1)

movie_reviews = pd.DataFrame({
    'review_content': review_content,
    'review_label':review_label
})
```

```python
movie_reviews.to_csv('movie_reviews.csv', index=False, encoding='utf-8')
```
